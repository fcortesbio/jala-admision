### Ejercicio 1 Ayudando a Sofía con su Casa Inteligente
Sofía acaba de instalar un sistema inteligente en su casa para controlar la temperatura. Pero necesita tu ayuda para que funcione correctamente. Sofía quiere que el sistema tome decisiones automáticas dependiendo de la temperatura:

- Si hace mucho calor (25°C o más): Sofía quiere que el sistema encienda el aire acondicionado.
-Si la temperatura está agradable (entre 18°C y 24°C): Sofía quiere que el sistema no haga nada, porque el clima es perfecto.
-Si hace frío (18°C o menos): Sofía necesita que el sistema encienda la calefacción. hace
 frío (temperatura menor o igual a 18°C), el sistema enciende la calefacción para mantener el calor.

### Ejercicio 2: La Fiesta Exclusiva de los Números
Susana es una jefa muy exigente, la cual esta organizando una gran fiesta sorpresa para los empleados de una empresa.
Pero no todos podran participar, cada empleado tiene un número asignado al azar que determina si puede entrar y qué nivel de acceso tendrá.
¡Tu trabajo es ayudarles a saber si podrán disfrutar de la fiesta y qué áreas estarán disponibles para ellos!

Aquí están las reglas que debes programar para que el sistema de acceso funcione:

* Par y múltiplo de 7:
Este número tiene Acceso VIP a una zona exclusiva, con los mejores servicios y privilegios.

* Par y múltiplo de 5:
Este número puede entrar al Área Principal de la fiesta, un espacio cómodo con buena música y comida.

* Par, múltiplo de 5 y 7:
Este número es muy especial, recibe Acceso a Todas las Áreas, convirtiéndose en el invitado más importante.

* Par, pero no múltiplo de 5 ni de 7:
Este número tiene Acceso Limitado, puede entrar, pero solo a ciertas zonas.

* Impar:
Este número no puede entrar a la fiesta, ya que no cumple con las reglas de acceso.

### Ejercicio 3: La Selección de Tareas en una Oficina
Imagina que trabajas en una oficina donde debes priorizar las tareas del día.
Tienes una lista de tareas y debes decidir en qué orden realizarlas.
Cada tarea tiene un nivel de prioridad y necesitas decidir si realizarla, posponerla o delegarla.

Condiciones:
- Si la tarea es de alta prioridad, debes hacerla inmediatamente.
- Si la tarea es de prioridad media, puedes posponerla para después.
- Si la tarea es de baja prioridad, puedes delegarla a otro compañero.

Tareas:
- Tarea 1: Reunión con el cliente (alta prioridad)
- Tarea 2: Responder correos electrónicos (prioridad media)
- Tarea 3: Organizar los archivos (baja prioridad)

### Ejercicio 4: El Alcalde y su Registro de Votantes
Marco, el alcalde de una ciudad, está organizando las próximas elecciones
y necesita asegurarse de que todas las personas que cumplan con la edad legal para
votar sean registradas correctamente.

Marco establece las siguientes reglas:
* Personas aptas para votar: Tienen al menos 18 años y no superan los 120 años (edad máxima razonable para un votante activo).
-Personas no aptas para votar: Son menores de 18 años, ya que aún no tienen la edad legal para votar.
-Errores en el registro: Si alguien ingresa un número menor o igual a 0, o mayor a 120, algo está mal y debe corregirse.


### Ejercicio 5: El Aventurero y la Puerta de las Decisiones
Imagina que estás creando un juego donde el jugador es un aventurero que acaba de llegar a la entrada de un mundo mágico. Frente a él, aparece una gran puerta con dos opciones:  
"Entrar" para iniciar su aventura, o "Salir" para abandonar la misión.

- Si el jugador decide entrar, un guardián aparece y le da a elegir entre diferentes capítulos,
 cada uno representando una parte emocionante de la aventura.
-Si el jugador decide salir, simplemente abandona el mundo mágico con un mensaje de despedida.
-Si ingresa algo que el guardián no entiende, se le pedirá que elija de nuevo para poder continuar.
