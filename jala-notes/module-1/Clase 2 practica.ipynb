**Ejercicio 6: Mi tienda de postres**

Imagina que tienes una tienda de postres. En tu men tienes 3 tipos de postres: pastel (10 dlares), galletas (5 dlares) y helado (3 dlares).
Un cliente llega y decide comprar una cantidad especfica de cada uno de estos postres.

Necesitamos calcular el total de la compra en funcin de la cantidad que el cliente elija.

Crea un programa que te ayude a resolver el problema.

```python
# Solucin modular completa
def calcular_total(carrito: Dict[str, int], catalogo: Dict[str, float])->float:
    """
    carrito: objeto con los productos y cantidades de la venta
    catalogo: objeto con la lista de productos y precios disponibles
    calcula y retorna el valor total de compra
    """
    
    if not carrito:
        return 0.0
    elif not catalogo:
        raise ValueError("Falta la lista de precios")
    
    total: float = 0.0
    
    for producto, cantidad in carrito.items():
        if producto not in catalogo:
            raise KeyError(f"Producto '{producto}' no encontrado en el catlogo")
        total += catalogo[producto] * cantidad
    return total

def construir_carrito(productos: list[str])-> Dict[str, int]:
    """
    Pregunta por cantidades usando la funcin input
    prompt: mensaje personalizado
    """
    carrito: Dict[str, int] = {}
    for producto in productos:
        carrito[producto] = prompt_int(f"Ingrese cantidad para producto: {producto}")        
    return carrito
    
def prompt_int(message: str= None)-> int:
    """
    helper to user-prompting a positive number
    """
    if not message: 
        message = "Enter a number: " 
    while True:
        try:
            number: int = int(input(message))
            if number >= 0: 
                return number
            raise(ValueError("La cantidad debe ser un nmero entero positivo."))
        except ValueError as e:
            print(e)


if __name__ == "__main__":
    postres = {
        "pastel": 10.0,
        "galletas": 5.0,
        "helado": 3.0
    }
       
    
    pre_order = ["pastel", "helado"]
    carrito_ejemplo = construir_carrito(["pastel", "helado"])
    print(carrito_ejemplo)
    # compra_ejemplo = {"pastel": 5, "galletas": 40}
    total = calcular_total(carrito_ejemplo, postres)
    print(f"El precio total es: {total} USD")
```

```python
# Solucin simple : No estructuras de control
precio_pastel, precio_galletas, precio_helado = 10.0, 5.0, 3.0
# cantidad_galletas, cantidad_helados = 5, 40, 0
cantidad_pasteles = prompt_int("Cuntos pasteles?")
cantidad_helados = prompt_int("Cuntos helados?")
cantidad_galletas = prompt_int("Cuntas galletas?")

total = precio_pastel*cantidad_pasteles + precio_galletas*cantidad_galletas + precio_helado*cantidad_helados 
print(f"El precio total es: {total} USD")
```

**Ejercicio 7: Clasificacin de edades**

Clasifica a un grupo de personas segn su edad. Considera que un persona es:

Un nio si tiene menos de 18 aos

Un adulto si tiene entre 18 aos menos de 65 aos

Un adulto mayor si tiene mas de 65 aos




```python
# Solucion simple : sin estructuras de control
edad: int = prompt_int("Ingrese su edad: ")

is_kid: bool = edad < 18
is_adult: bool = 18 <= edad <= 65
is_senior: bool = edad > 65

print("Usted es un nio?: ", is_kid)
print("Usted es un adulto?: ", is_adult)
print("Usted es un adulto mayor?: ", is_senior)
```

```python
# solucin modular

def clasificar_edad(edad: int)-> str:
    if edad > 65:
        return "senior"
    elif edad >= 18: 
        return "adult"
    else:
        return "kid" 
 

def print_category(category: str)-> None:
    match category:
        case "kid":
            print("Usted es un nio.")
        case "adult":
            print("Usted es un adulto.")
        case "senior":
            print("Usted es un adulto mayor.")
        
if __name__ == "__main__":
    cat = clasificar_edad(prompt_int("Edad?"));
    print_category(cat)
```

**Ejercicio 8: Adivina el nmero**

Crea un juego donde el usuario debe adivinar un nmero aleatorio entre 1 y 10 .


```python
from random import randint
from IPython.display import clear_output
from time import sleep

def adivina(max: int):
    magic_number: int = randint(1, max)
    choices: list = []
    attempts: int = 0
    print(magic_number) # borrar --debug--
    
    while True:
        print("Puedes adivinar el nmero mgico?")
        guess = get_capped_int(f"Ingresa un nmero entre 1 y {max}", max)
        if guess == magic_number:
            attempts +=1
            print(f"Fecilidades!, el nmero mgico es: {magic_number}")
            print(f"Nmero de intentos: {attempts}")
            break
        clear_output()
        print("Oops, not this time. Let's try again")
        choices.append(guess)
        print("Your choices: ", choices)
        attempts += 1
        print(f"Intentos: {attempts}")
        sleep(1)

def get_capped_int(message: str, max: int = 10)-> int:
    """
    helper del helper 
    prompt_int valida si el user input es positivo
    este helper agrega funcionalidad de un lmite superior (max)
    """ 
    while True:
        number = prompt_int(message or "Ingresa un nmero: ")
        if 0 < number <= max:
            return number
   
        
if __name__ == "__main__":
    adivina(10)
```

**Ejercicio 9: Qu hora es?**

Juan est estudiando muy duro para las pruebas de Jala University y quiere saber cuando puede tomar un descanso.

El dijo que descansara a partir de las 14:00

Ayuda a Juan a saber si es hora de un descanso.


```python
from datetime import datetime as dt

def is_nap_time()->bool:
    now = dt.now()
    # print(now)
    # print(now.hour)
    return True if now.hour >= 14 else False
    
if __name__ == "__main__":
    print(is_nap_time())
```
