### Ejercicio 1: Una Gran Fiesta
Jose y Valeria una pareja de feliz casados, son los principales encargados
de la organizacion de una fiesta planeada por toda su familia
es un gran evento familiar, por tanto cada miembro de la
familia de Jose y Valeria tiene la obligacion de llevar ciertos artículos y registrarse. Cada miembro debe llevar al menos 3 artículos para ayudar con la preparación del evento. Si un miembro lleva más de 5 artículos, se considera que ha ayudado excepcionalmente.  
Tambien, la feliz pareja desearia saber quienes son los familiares que han ayudado excepcionalmente.

Para esta tarea la pareja necesita ayuda con un programa. Valeria argumenta que seria buena idea hacer un menu 3 opciones,
Jose pide que la primera opcion sirva para registrar al familiar y saber cuantos articulos trajo.
Para la segunda opcion les gustaria que se imprima en pantalla
quienes fueron los familiares que ayudaron excepcionalmente.
Y finalmente la tercera opcion para apagar el programa
El programa no debe apagarse hasta que se elija la tercera opcion

Podremos ayudar a esta feliz pareja?

Condiciones:
* Si un miembro lleva más de 5 artículos, se considera que ha ayudado excepcionalmente.
* Si lleva entre 3 y 5 artículos, ha ayudado, pero sin destacar tanto.
* Si lleva menos de 3 artículos, no ha ayudado lo suficiente.
* Junto al mensaje debe de imprimirse el nombre del familiar ingresado.


### Ejercicio 2: La Tienda De Martha
Martha acaba de abrir una pequeña tienda de barrio donde
vende productos básicos como frutas, pan y leche.
Sin embargo, a veces le cuesta llevar un control de su inventario.
A veces se olvida de qué productos tiene disponibles,
o sus clientes le piden algo y no sabe si todavía le queda.

Martha necesita nuestra ayuda para organizar su tienda.
Ayudémosla creando un programa que le permita:

* Ver qué productos tiene en su tienda.
-Agregar nuevos productos al inventario cuando recibe más mercancía.
-Buscar rápidamente si un producto está disponible cuando un cliente lo pide.
-Salir del sistema cuando ya no lo necesite.


### Ejercicio 3: Puntajes De Un Examen Misterioso
Valentina es una maestra de matematicas que esta organizando una prueba misteriosa
El objetivo de la prueba es verificar el nivel de rendimiento de toda su clase,
para mantener el anonimato y misterio de la prueba, los estudiantes no necesitan colocar nombre a la prueba
por lo tanto Valentina solo tendra una lista de puntajes.

Para saber el rendimiento de su curso, se necesita un promedio de todas las notas que obtengan sus estudiantes
imprimiendo en pantalla la cantidad de notas analizadas, el promedio y un mensaje que avise si el curso aprobo o no.

Valentina tambien quisiera que antes de mostrar el promedio del curso,se muestre en pantalla todas las
notas obtenidas por sus estudiantes, junto con un mensaje (dependiendo de la nota).

La impresion de cada nota una a una, mantendra el suspenso de saber si el curso aprobo o no.
Podemos ayudar a Valentina?

Condiciones:
* Si un puntaje tiene 90 o más puntos, el mensaje es "Felicidades, esto ayuda al promedio del curso! :D".
* Si un puntaje es menor a 90 y mayor o igual a 70, el mensaje es "Excelente, vamos por buen camino :D".
* Si un puntaje es menor a 70 y mayor o igual a 51, el mensaje es "Buen trabajo :)".
* Si un puntaje es menor a 51, el mensaje es "Debemos esforzarnos un poco mas, el promedio peligra D:"
* Si el promedio es mayor o igual a 51 se imprime "El curso aprobo!" de lo contario, se imprime "El curso no aprobo..."

### Ejercicio 4
Carlos es un estudiante que está aprendiendo a programar y, al mismo tiempo,
le gusta ayudar a sus amigos mas pequeños con sus tareas de matemáticas.
Ya que, a veces se les hace difícil recordar todas las
operaciones que tienen que realizar y calcular rápidamente los resultados.

Para solucionar este problema, Carlos quiere un programa
que funcione como una herramienta sencilla para hacer cálculos básicos.

Ayudémoslo creando un programa que muestre un menu que:
- Le permita realizar operaciones matemáticas básicas como sumar o restar.
- Le dé la opción de salir del programa cuando ya no necesite calcular más.
- Se asegure de que elija opciones válidas para no confundirlo.

tomando todas las flores, se calcula con el factorial de n.
